```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
from wordcloud import WordCloud,STOPWORDS
from nltk.corpus import stopwords
```

```python
import warnings
warnings.filterwarnings(action='ignore', category=DeprecationWarning)
```

```python
review_df=pd.read_csv("hospital_review_afterpreprocessed.csv", index_col=[0])
```

### Distribution of ratings

```python
review_df['star'].value_counts().plot.pie(figsize=(20,10),autopct='%1.1f%%', startangle=270, fontsize=15);
```

### WordClouds

```python
preprocessed_text = " ".join(word for word in review_df.review_preprocessed if word != "nan")
```

```python
def wc_visualize(preprocessed_text):    
        wordcloud_preprocessed=WordCloud(width=800, height=600, background_color='white', stopwords=stopwords.words('english')).generate(preprocessed_text)
        plt.figure(figsize=(12,16), facecolor=None)
        plt.imshow(wordcloud_preprocessed)
        plt.axis("off")
        plt.tight_layout(pad=0)
        plt.show()
```

### WordClouds (All)

```python
wc_visualize(preprocessed_text)
```

### WordClouds (Positive)

```python
preprocessed_text_positive= " ".join(word for word in review_df[review_df['star']>=3].review_preprocessed if word != "nan")
wc_visualize(preprocessed_text_positive)
```

#### Word Clouds (Negative Review)

```python
preprocessed_text_negative= " ".join(word for word in review_df[review_df['star']<3].review_preprocessed if word != "nan")
wc_visualize(preprocessed_text_negative)
```
